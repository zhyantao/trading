```python
import akquant as aq
import akshare as ak
from akquant import Strategy

# 1. 准备数据
# 使用 akshare 获取 A 股历史数据 (需安装: pip install akshare)
df = ak.stock_zh_a_daily(symbol="sh600000", start_date="20260101", end_date="20260308")


class MyStrategy(Strategy):
    def on_bar(self, bar):
        # 简单策略示例:
        # 当收盘价 > 开盘价 (阳线) -> 买入
        # 当收盘价 < 开盘价 (阴线) -> 卖出

        # 获取当前持仓
        current_pos = self.get_position(bar.symbol)

        if current_pos == 0 and bar.close > bar.open:
            self.buy(bar.symbol, 100)
            print(f"[{bar.timestamp_str}] Buy 100 at {bar.close:.2f}")

        elif current_pos > 0 and bar.close < bar.open:
            self.close_position(bar.symbol)
            print(f"[{bar.timestamp_str}] Sell 100 at {bar.close:.2f}")


# 运行回测
result = aq.run_backtest(
    data=df,
    strategy=MyStrategy,
    symbol="sh600000"
)

# 打印回测结果
print("\n=== Backtest Result ===")
print(result.metrics_df)
```

```python
"""
东方财富K线数据获取 - 使用 curl_cffi 绕过反爬虫
curl_cffi 能够模拟浏览器的 TLS/JA3 指纹，有效绕过基于指纹检测的反爬机制
"""

from curl_cffi import requests  # 关键：使用 curl_cffi 替代标准 requests
import os
import time
import random
from typing import Optional, Dict, Any, List


# ============ 基础配置 ============

# 彻底禁用代理
os.environ['NO_PROXY'] = 'push2his.eastmoney.com,quote.eastmoney.com,eastmoney.com'
os.environ['no_proxy'] = 'push2his.eastmoney.com,quote.eastmoney.com,eastmoney.com'


# ============ 核心客户端类 ============

class EastMoneyClient:
    """
    东方财富数据客户端（使用 curl_cffi 绕过反爬）
    
    特性：
    1. 模拟 Chrome 120 的 TLS/JA3 指纹
    2. 自动预热会话获取 Cookie
    3. 智能重试和随机延迟
    4. 完整的请求头模拟
    """
    
    def __init__(self, impersonate: str = "chrome120", timeout: int = 15):
        """
        初始化客户端
        
        Args:
            impersonate: 模拟的浏览器类型，可选 chrome, chrome110, chrome120, edge, safari, firefox
            timeout: 请求超时时间（秒）
        """
        # 创建模拟浏览器指纹的会话 - 这是绕过反爬的关键！
        # curl_cffi 会模拟真实浏览器的 TLS 指纹、HTTP/2 行为等
        self.session = requests.Session(impersonate=impersonate)
        self.timeout = timeout
        self.base_url = "https://push2his.eastmoney.com"
        self.warmed_up = False
        
    def _get_headers(self) -> Dict[str, str]:
        """生成完整的浏览器请求头"""
        return {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
            'Accept': '*/*',
            'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8',
            'Accept-Encoding': 'gzip, deflate, br',
            'Referer': 'https://quote.eastmoney.com/',  # 关键：必须来自主站
            'Origin': 'https://quote.eastmoney.com',
            'Connection': 'keep-alive',
            'sec-ch-ua': '"Not_A Brand";v="8", "Chromium";v="120", "Google Chrome";v="120"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'Sec-Fetch-Dest': 'empty',
            'Sec-Fetch-Mode': 'cors',
            'Sec-Fetch-Site': 'same-site',
        }
    
    def _warmup(self) -> bool:
        """
        预热会话：访问主站建立 Cookie 和会话状态
        模拟真实用户先访问页面再请求数据的行为
        """
        if self.warmed_up:
            return True
            
        try:
            # 访问行情页面建立基础会话
            resp = self.session.get(
                'https://quote.eastmoney.com/concept/sh600519.html',
                headers=self._get_headers(),
                timeout=10
            )
            
            if resp.status_code == 200:
                self.warmed_up = True
                # 随机延迟，模拟人类操作间隔
                time.sleep(random.uniform(0.3, 0.8))
                return True
            return False
            
        except Exception as e:
            print(f"[Warmup Warning] {e}")
            # 预热失败不阻断，继续尝试主请求
            return False
    
    def _safe_request(self, url: str, params: Dict, max_retries: int = 3) -> Optional[Any]:
        """
        带重试机制的安全请求
        
        Args:
            url: 请求URL
            params: 查询参数
            max_retries: 最大重试次数
        """
        # 确保已预热
        if not self._warmup():
            print("[Warning] 会话预热失败，继续尝试...")
        
        for attempt in range(max_retries):
            try:
                # 指数退避 + 随机抖动
                if attempt > 0:
                    sleep_time = (2 ** attempt) + random.uniform(0, 2)
                    print(f"[Retry {attempt}/{max_retries}] 等待 {sleep_time:.1f}s 后重试...")
                    time.sleep(sleep_time)
                else:
                    # 首次请求也添加随机延迟，避免请求过于规律
                    time.sleep(random.uniform(0.5, 1.5))
                
                response = self.session.get(
                    url,
                    params=params,
                    headers=self._get_headers(),
                    timeout=self.timeout
                )
                
                if response.status_code == 200:
                    return response.json()
                elif response.status_code == 403:
                    print(f"[Error] 403 Forbidden - 可能被反爬拦截")
                    return None
                else:
                    print(f"[HTTP Error] Status: {response.status_code}")
                    
            except requests.errors.ConnectionError as e:
                print(f"[Connection Error] 连接被重置 (attempt {attempt+1}): {e}")
            except requests.errors.Timeout:
                print(f"[Timeout] 请求超时 (attempt {attempt+1})")
            except Exception as e:
                print(f"[Request Error] {type(e).__name__}: {e}")
        
        return None

    def get_kline(
        self,
        symbol: str,
        secid: Optional[str] = None,
        period: str = "daily",
        start_date: Optional[str] = None,
        end_date: Optional[str] = None,
        fq: int = 1
    ) -> Optional[Dict[str, Any]]:
        """
        获取K线数据
        
        Args:
            symbol: 股票代码，如 "600519"
            secid: 东方财富格式，如 "1.600519"（自动转换）
            period: 周期 daily/weekly/monthly
            start_date: 开始日期 YYYYMMDD，默认180天前
            end_date: 结束日期 YYYYMMDD，默认今天
            fq: 复权类型 1前复权/0不复权/2后复权
            
        Returns:
            原始API返回的JSON字典
        """
        # 自动转换 secid 格式
        if secid is None:
            if symbol.startswith('6'):
                secid = f"1.{symbol}"  # 上海
            elif symbol.startswith('0') or symbol.startswith('3'):
                secid = f"0.{symbol}"  # 深圳主板/创业板
            elif symbol.startswith('688'):
                secid = f"1.{symbol}"  # 科创板
            elif symbol.startswith('8') or symbol.startswith('4'):
                secid = f"0.{symbol}"  # 北交所/新三板
            else:
                secid = f"0.{symbol}"
        
        # 默认日期范围
        if end_date is None:
            end_date = time.strftime('%Y%m%d')
        if start_date is None:
            from datetime import datetime, timedelta
            start_date = (datetime.now() - timedelta(days=180)).strftime('%Y%m%d')
        
        # 周期映射
        klt_map = {'daily': 101, 'weekly': 102, 'monthly': 103}
        klt = klt_map.get(period, 101)
        
        params = {
            'ut': '7eea3edcaed734bea9cbfc24409ed989',
            'fields1': 'f1,f2,f3,f4,f5,f6',
            'fields2': 'f51,f52,f53,f54,f55,f56,f57,f58,f59,f60,f61',
            'klt': klt,
            'fqt': fq,
            'secid': secid,
            'beg': start_date,
            'end': end_date
        }
        
        return self._safe_request(
            f"{self.base_url}/api/qt/stock/kline/get",
            params
        )
    
    def parse_klines(self, data: Dict) -> List[Dict]:
        """
        解析K线数据为结构化字典列表
        
        Args:
            data: API返回的原始数据
            
        Returns:
            解析后的K线列表，每条包含 date, open, close, high, low, volume 等字段
        """
        if not data or not data.get('data') or not data['data'].get('klines'):
            return []
        
        records = []
        for line in data['data']['klines']:
            fields = line.split(',')
            if len(fields) >= 9:
                records.append({
                    'date': fields[0],
                    'open': float(fields[1]) if fields[1] != '-' else None,
                    'close': float(fields[2]) if fields[2] != '-' else None,
                    'high': float(fields[3]) if fields[3] != '-' else None,
                    'low': float(fields[4]) if fields[4] != '-' else None,
                    'volume': float(fields[5]) if fields[5] != '-' else None,
                    'amount': float(fields[6]) if len(fields) > 6 and fields[6] != '-' else None,
                    'amplitude': float(fields[7]) if len(fields) > 7 and fields[7] != '-' else None,
                    'change_pct': float(fields[8]) if len(fields) > 8 and fields[8] != '-' else None,
                    'change_amount': float(fields[9]) if len(fields) > 9 and fields[9] != '-' else None,
                    'turnover': float(fields[10]) if len(fields) > 10 and fields[10] != '-' else None,
                })
        return records


# ============ 使用示例 ============

if __name__ == "__main__":
    # 创建客户端（自动模拟 Chrome 浏览器指纹）
    client = EastMoneyClient(impersonate="chrome120")
    
    print("=" * 70)
    print("示例1: 获取贵州茅台(600519)最近日线数据")
    print("=" * 70)
    
    # 获取数据
    raw_data = client.get_kline(
        symbol="600519",
        start_date="20250101",
        end_date="20250308"
    )
    
    if raw_data:
        # 解析数据
        klines = client.parse_klines(raw_data)
        stock_name = raw_data['data'].get('name', 'Unknown')
        
        print(f"\n股票名称: {stock_name}")
        print(f"数据条数: {len(klines)}")
        print(f"\n最近5个交易日:")
        print("-" * 70)
        
        for k in klines[-5:]:
            print(f"{k['date']} | "
                  f"开盘: {k['open']:>10.2f} | "
                  f"收盘: {k['close']:>10.2f} | "
                  f"涨跌: {k['change_pct']:>+6.2f}% | "
                  f"成交量: {k['volume']:>12.0f}")
    
    print("\n" + "=" * 70)
    print("示例2: 获取宁德时代(300750)周线数据")
    print("=" * 70)
    
    raw_data2 = client.get_kline(
        symbol="300750",
        period="weekly",
        start_date="20240101",
        end_date="20250308"
    )
    
    if raw_data2:
        klines2 = client.parse_klines(raw_data2)
        print(f"\n获取到 {len(klines2)} 周数据")
        print(f"最近3周:")
        for k in klines2[-3:]:
            print(f"{k['date']} | 收盘: {k['close']:.2f} | 周涨跌: {k['change_pct']:+.2f}%")
```
